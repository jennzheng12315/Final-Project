# The Metric:
Our metric explores which neighborhood has is the best in terms of a well-rounded life. To determine this, we use datasets that show the median death age for each neighborhood, information on playgrounds, and data on street art in each neighborhood. For the playground data set, we took into account the number of playgrounds. Another dataset was used to show the median death age per neighborhood, and our final data set showed how much art was in a given neighborhood. From there, we gave each neighborhood a ranking out of ten which we found from dividing the total art, playgrounds, and the median death age by the greatest number in the dataset and multiplied that by ten. We then combined the ratings and found the average rating for each neighborhood which is how we measured our metric.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas as geo
hood=geo.read_file("Neighborhoods/Neighborhoods_.shp")

play = pd.read_csv("https://data.wprdc.org/datastore/dump/e39ef76e-0a11-47c8-a86f-a37f55db7a2b")
public_art_data = pd.read_csv("public-art.csv", index_col="id")
death = pd.read_csv("https://data.wprdc.org/dataset/ee735209-4de7-4ea4-b446-bf0f0f6d9cb3/resource/c2e1500a-a12a-4e91-be94-76c6a892b7e2/download/nhoodmedianage20112015.csv",parse_dates=True)
```

## Playground Visualization

```python
neighbor=play.groupby('neighborhood').count()
play_map = hood.merge(neighbor, how='left', left_on='hood', right_on='neighborhood')
play_map.plot(column='ada_accessible',
             cmap='cool',              # choose a color palette
               edgecolor="white",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Playgrounds"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
             )
```

```python
neighborhood=play['neighborhood'].value_counts().sort_values(ascending=False).reset_index()

#reset=neighborhood.reset_index()

death10=death.sort_values(by=["TOTAL MD AGE AT DEATH"], ascending=False)
dropped=death10.drop(columns=['BLACKdeaths','Black MD AGE AT DEATH', 'WHITEdeaths','White MD AGE AT DEATH','TOTALdeaths*']).reset_index()
neighborhoods = public_art_data['neighborhood'].value_counts().sort_values(ascending=False).reset_index()
dropped.head()
```

```python
#neighborhood['neighborhood']=neighborhood['neighborhood'].str.lower()

play_rate=pd.DataFrame(neighborhood)
#highest=rate.max()
#print(highest)
play_rate["Rating"]=((play_rate['neighborhood']/32)*10)
#play_rate["Number of Playgrounds"]=(play_rate['neighborhood'])
#play_rate["Neighborhood"](play_rate[])
dropped['NEIGHBORHOOD']=dropped['NEIGHBORHOOD'].str.lower()
death_rate=pd.DataFrame(dropped)
#rate.groupby("NEIGHBORHOOD")
#highest=rate.max()
#print(highest)
death_rate["Rating"]=((death_rate['TOTAL MD AGE AT DEATH']/85.9)*10)

#neighborhoods['neighborhood']=neighborhoods['neighborhood'].str.lower()

art_rate=pd.DataFrame(neighborhoods)
#highest=rate.max()
#print(highest)
art_rate["Rating"]=((art_rate['neighborhood']/27)*10)

#art_rate["id"]=()
art_rate.head()
```

```python
metric=play_rate.merge(art_rate, how='left', on='index')
final_metric=metric.merge(death_rate, how='left', left_on='index', right_on='NEIGHBORHOOD')
#death_rate['neighborhood']=death_rate['NEIGHBORHOOD']
#final_metric=metric.merge(death_rate, how='left', on='neighborhood')
final_metric.head()

#rate_metric=play_rate.merge(art_rate.merge(death_rate, how='left', left_on='index', right_on='NEIGHBORHOOD'), how='left', left_on='index', right_on='index')
#rate_metric.head()
```
